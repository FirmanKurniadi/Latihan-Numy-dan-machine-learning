```python
import matlotlib.pylot as plt
import numpy as np
```

```python
x = np.linspace(1,5,11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
plt.plot(x, y, 'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Data Contoh')
plt.show() # bisa digunakan peganti  shift enter
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'b*-')
# membuat 2 lot dalam 1 canvas
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1 , 0.1 , 1 , 2])

axes.plot(x,y,'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title( 'Contoh Judul')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1 , 0.1 , 0.8 , 0.8])
axes2 = fig.add_axes([0.2 , 0.5 , 0.4 , 0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('x utama')
axes1.set_ylabel('y utama')
axes1.set_title('plot utama')

axes2.plot(x,y,'r')
axes2.set_xlabel('x dalam')
axes2.set_ylabel('y dalam')
axes2.set_title('contoh judul')
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8,4), dpi=100)

axes.plot(x,y,'r')
axes.set_xlabel('x ')
axes.set_ylabel('y ')
axes.set_title('plot judul')

fig.savefig("contoh.png",dpi=100)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x** 2, label="Data x pangkat 2")
axes.plot(x, x** 3, label="Data x pangkat 3")

axes.legend(loc=0)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color="red", linewidth=0.25)
axes.plot(x, x+2, color="red", linewidth=0.50)
axes.plot(x, x+3, color="red", linewidth=0.75)
axes.plot(x, x+4, color="yellow", linewidth=1)
#style plot
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color="blue", lw=3, linestyle='-')
axes.plot(x, x+2, color="blue", lw=3, ls='-')
axes.plot(x, x+3, color="blue", lw=3, ls=':')
axes.plot(x, x+4, color="blue", lw=3, linestyle='--')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color="blue", lw=3, ls='-', marker='+')
axes.plot(x, x+2, color="blue", lw=3, ls='-', marker='o')
axes.plot(x, x+3, color="blue", lw=3, ls='-', marker='s')

#symbol2nya terdapat di matplotlib.org/3.11/api/markers_api.html
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

line, = axes.plot(x, x+1, color="blue", lw=1.50, )
line.set_dashes([5,5,15,5])
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

line, = axes.plot(x, x+1, color="blue", lw=1.50, )
line.set_dashes([10,5,3,5])
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x+1, color="blue", lw=3, ls='-', marker='+', markersize='10')
axes.plot(x, x+2, color="blue", lw=3, ls='-', marker='o', markerfacecolor='red')
axes.plot(x, x+3, color="blue", lw=3, ls='-', marker='s', markerfacecolor='yellow', markeredgewidth=3,markeredgecolor="green")
```
